## Title: Final Project first year Student Performance Analysis


# Step 1: Setup and File Upload



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Rename columns for easier typing and standard formatting
df.rename(columns={
    'race/ethnicity': 'race_ethnicity',
    'parental level of education': 'parent_education',
    'test preparation course': 'test_prep',
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score',
    'Average Score': 'average_score'
}, inplace=True)

print("Dataset loaded successfully!")
print(df)
```

# Step 2 – Basic Dataset Inspection

```python
# Checking the overall shape, data types, and first few rows to understand the dataset structure.
print(f"Shape of dataset: {df.shape}\n")
print("--- First 5 Rows ---")
display(df.head())
print("\n--- Last 5 Rows ---")
display(df.tail())
print("\n--- Data Types ---")
print(df.dtypes)
print("\n--- Dataset Info ---")
df.info()
print(f"\nColumns: {list(df.columns)}")
```

# Step 3 – Descriptive Statistics


```python
# Generating basic statistical measures like mean, median, and standard deviation for numerical columns.
print("--- Numerical Summary ---")
display(df.describe())

print("\n--- Individual Statistics ---")
print(f"Mean Math Score: {df['math_score'].mean():.2f}")
print(f"Median Reading Score: {df['reading_score'].median()}")
print(f"Std Dev of Writing Score: {df['writing_score'].std():.2f}")
```

# Step 4 – Missing Value Analysis

```python
# Check for nulls
missing_data = df.isnull().sum()
print("--- Missing Values Count ---")
print(missing_data)

# Check percentage of missing values
missing_percent = (df.isnull().sum() / len(df)) * 100
print("\n--- Missing Values Percentage ---")
print(missing_percent)

# If there were missing values, you could drop them using df.dropna(inplace=True)
```

# Step 5 – Frequency Counts (value_counts)


```python
# Counting the occurrences of each distinct category within our categorical variables.
categorical_cols = ['gender', 'race_ethnicity', 'parent_education', 'lunch', 'test_prep']

for col in categorical_cols:
    print(f"\n--- Frequency for {col} ---")
    print(df[col].value_counts())
```

# Step 6 – Unique Values (unique & nunique)

```python
# Identifying the exact unique categories present in each column to understand data variety.
print("--- Number of Unique Values per Column ---")
print(df.nunique())

print("\n--- Unique Values in Categorical Columns ---")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

# Step 7 – Percentage Distribution

```python
for col in categorical_cols:
    print(f"\n--- Percentage Distribution for {col} ---")
    print(round(df[col].value_counts(normalize=True) * 100, 2).astype(str) + '%')
```

# Step 8 – Filtering Data

```python
# 1. Students who completed test prep course
prep_completed = df[df['test_prep'] == 'completed']

# 2. Students on free/reduced lunch
free_lunch = df[df['lunch'] == 'free/reduced']

# 3. Students scoring above 90 average
top_students = df[df['average_score'] >= 90]

# 4. Students with math score < 40 (at-risk)
math_risk = df[df['math_score'] < 40]

# 5. Female students from group E
female_group_E = df[(df['gender'] == 'female') & (df['race_ethnicity'] == 'group E')]

print(f"Number of top students (Avg >= 90): {len(top_students)}")
print(f"Number of at-risk math students: {len(math_risk)}")
```

# Step 9 – Cross Tabulation (pd.crosstab)

```python
print("--- Gender vs Test Prep ---")
display(pd.crosstab(df['gender'], df['test_prep']))

print("\n--- Race/Ethnicity vs Lunch ---")
display(pd.crosstab(df['race_ethnicity'], df['lunch']))

print("\n--- Parent Education vs Test Prep ---")
display(pd.crosstab(df['parent_education'], df['test_prep']))
```

# Step 10 – Grouping Data (groupby)

```python
# Mean scores grouped by gender
print("--- Mean Scores by Gender ---")
display(df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean())

# Max/min scores by race/ethnicity
print("\n--- Max & Min Math Scores by Race/Ethnicity ---")
display(df.groupby('race_ethnicity')['math_score'].agg(['max', 'min']))
```

# Step 11 – Sorting

```python
# Sort by Average Score descending (Top 5 performers)
top_performers = df.sort_values(by='average_score', ascending=False).head()
display(top_performers)

# Sort by Math Score ascending (Bottom 5 students)
struggling_math = df.sort_values(by='math_score', ascending=True).head()
display(struggling_math)
```

# Step 12 – Adding Derived Columns



```python
# 1. Grade Category
def assign_grade(score):
    if score >= 80: return 'Distinction'
    elif score >= 60: return 'First Class'
    elif score >= 40: return 'Pass'
    else: return 'Fail'

df['grade_category'] = df['average_score'].apply(assign_grade)

# 2. Score Gap (Reading - Math)
df['score_gap'] = df['reading_score'] - df['math_score']

print("--- Data with New Derived Columns ---")
display(df[['math_score', 'reading_score', 'average_score', 'grade_category', 'score_gap']].head())
```

```python
# 13 Radar Plot (Spider Chart) → Average Subject Scores Comparison (Male vs Female)
import numpy as np

categories = ['Math', 'Reading', 'Writing']
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # Close the loop

male_scores = df[df['gender'] == 'male'][['math_score', 'reading_score', 'writing_score']].mean().tolist()
female_scores = df[df['gender'] == 'female'][['math_score', 'reading_score', 'writing_score']].mean().tolist()
male_scores += male_scores[:1]
female_scores += female_scores[:1]

ax = plt.subplot(111, polar=True)
ax.plot(angles, male_scores, 'o-', linewidth=2, label='Male')
ax.fill(angles, male_scores, alpha=0.2)
ax.plot(angles, female_scores, 'o-', linewidth=2, label='Female')
ax.fill(angles, female_scores, alpha=0.2)
ax.set_thetagrids(np.degrees(angles[:-1]), categories)
ax.set_title('Average Subject Scores Comparison')
ax.legend(loc='upper right')
plt.show()
```

```python
# 9 Average Scores by Parental Level of Education
avg_scores_by_parent_edu = df.groupby('parent_education')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()

avg_scores_melted = avg_scores_by_parent_edu.melt(
    id_vars=['parent_education'],
    value_vars=['math_score', 'reading_score', 'writing_score'],
    var_name='Subject',
    value_name='Average Score'
)

plt.figure(figsize=(12, 7))
sns.barplot(
    x='parent_education',
    y='Average Score',
    hue='Subject',
    data=avg_scores_melted,
    palette='viridis'
)
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Subject')
plt.tight_layout()
plt.show()
```
